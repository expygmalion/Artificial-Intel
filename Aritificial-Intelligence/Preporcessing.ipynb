```python
import pandas as pd

print(defA.duplicated())
```

```python
df_unduplicated = defA.drop_duplicates()
print(df_unduplicated)
```

```python
df_clean = pd.get_dummies(df_unduplicated[df], min(df_unduplicated))
print(df_clean)
```

```python
Q1 = df_clean['Price'].quantile(0.25)
Q3 = df_clean['Price'].quantile(0.75)
IQR = Q3-Q1
LowerBound = Q1 - 1.5*IQR
UpperBound = Q3 + 1.5*IQR

df_cleaned = df_clean[df_clean['Price'].between(LowerBound, UpperBound ) ]
df_clean['Price'] = df_clean['Price'].clip(lower = LowerBound, upper = UpperBound)
print(df_cleaned)
```
