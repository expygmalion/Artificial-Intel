```python
import pandas as pd

df = pd.read_csv("Projects/PythonProjects/Try 1/car_sales - original.csv")
print(df["Make"])
print(df.iloc[0:10])
print(df.iloc[0: 3])
```

```python
## apply grouping: 

print(df.groupby(["Make", "Colour"]).sum())
print("\n--------------------------------------------")
print(df.groupby(['Make', "Colour"]).max())
```

```python
print(df[["Price"]].mode())
print(df.groupby(["Colour","Make"]).agg({"Price": ['sum', 'max'], "Doors": 'max'}))
```
